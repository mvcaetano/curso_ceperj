### Formas de se usar a função **print**

````python
# Vamos declarar duas variáveis

nome_inicial  = 'The Rolling'
nome_final = ' Stones'
````

Uma maneira de passar o nome da banda para a função print (mais antiga)

````python
print('Uma banda que curto é %s%s' %(nome_inicial, nome_final))
````

Outra maneira de passar o nome da banda para a função print

````python
print('O nome da banda que eu gosto é {}{}'.format(nome_inicial, nome_final))
````

````python
print('O nome da banda que eu gosto é {1}{0}'.format(nome_inicial, nome_final))
````

f-strings em Python

````python
print(f'O nome da banda que eu gosto é {nome_inicial}{nome_final}')
````

````python
# Usando f-strings

x = 10

print(f'O nome da banda que eu gosto é {x}')
````

````python
# Sem f-strings

x = 10

print('O nome da banda que eu gosto é {x}')
````
